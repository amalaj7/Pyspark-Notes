Rdd - Reselient Distributed Dataset
An RDD in Spark is simply an immutable distributed collections of objects. Each RDD IS split into multiple
partitions, which may be computed on different nodes of the cluster.

RDD can be created in two ways: by loading an external dataset, or by transforming one RDD into another.

```python
num = sc.parallelize([5,5,4,3,2,9,2],2)
num.collect()
```

```python
num.glom().collect()
```

```python
num.reduce(lambda a,b: a+b)
#reduce() is useful when you need to apply a function to an iterable and reduce it to a single cumulative value. 
```

```python
num.reduce(lambda a,b: a*b)
# the reduce() function receives two arguments, a function and an iterable.
#However, it doesn't return another iterable, instead it returns a single value. 
```

```python
# fold: the initial value for the accumulated result of each partition for the op operator, 
# and also the initial value for the combine results from different partitions
```

```python
num.fold(2,lambda a,b:a+b) # for fold it will take some initial value
```

```python
num.fold(2,lambda a,b : a*b ) # when you are adding 1 it becomes 32
# when you are adding 2 it becomes 34.
# when you are adding 3 it becomes 36
#by deafult sc.parallelize takes 1 partition, but it has 2 brackets , i will add 2
#if sc.parallelize takes 2 partition, but it has 3brackets ie one for whole,and other for 2 partitions(ie 3),i will add 3
# hence fold will be 36 when fold is set to 2 (2*3)
```

```python
from operator import add
b=sc.parallelize([1, 2, 3, 4, 5])
b.fold(1, add)
```

```python
from operator import add,mul
num3 = sc.parallelize([5,5,4,3,2,9,2]).fold(10,mul)
num3
```
